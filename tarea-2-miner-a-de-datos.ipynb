```python
import pandas as ps

covid_file_path = '/kaggle/input/covid-world-vaccination-progress/country_vaccinations.csv'
df = pd.read_csv(covid_file_path)

df.tail(10) #Mostrar las primeras 10 filas
```

```python
#ESTADISTICA DESCRIPTIVA

df.info() #Informacion general de las columnas
print("\n")

#verificar valores nulos
df.isnull().sum()
print("\n")

#media, desviación estándar, valores mínimos y máximos de las columnas numéricas.
df.describe()
print("\n")
```

```python
#Resumen de variables categoricas.
print(df.describe(include=["O"]))  # Incluye datos de tipo objeto (alfanuméricos)
```

```python
#AGRUPAR LOS DATOS POR ENTIDADES

# Agrupar por país y calcular estadísticas
df_grouped = df.groupby("country")[["total_vaccinations", "people_vaccinated", "people_fully_vaccinated", "daily_vaccinations"]].agg(["sum", "mean", "max", "min"])

# Mostrar los primeros resultados
print(df_grouped.head())
```

```python
# País con más vacunaciones totales
top_pais = df_grouped["total_vaccinations"]["sum"].idxmax()
print(f"País con más vacunaciones: {top_pais}")
print("\n")

# País con menos vacunaciones totales
bottom_pais = df_grouped["total_vaccinations"]["sum"].idxmin()
print(f"País con menos vacunaciones: {bottom_pais}")
print("\n")

# Promedio de vacunaciones diarias por país
df_daily_avg = df.groupby("country")[["daily_vaccinations"]].mean()
print(df_daily_avg.sort_values("daily_vaccinations", ascending=False).head(10))
print("\n")

# Contar cuántos registros tiene cada país
df_days_reported = df["country"].value_counts()
print(df_days_reported.head(10))
print("\n")
```
