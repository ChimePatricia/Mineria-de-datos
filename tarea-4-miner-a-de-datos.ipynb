```python
import pandas as pd
from scipy.stats import f_oneway

df = pd.read_csv('/kaggle/input/covid-world-vaccination-progress/country_vaccinations.csv') # Cargar los datos
df_validos = df[['country', 'daily_vaccinations']].dropna() # Filtrar columnas necesarias y eliminar filas con valores nulos
cantidad_registros = df_validos['country'].value_counts() # Contar registros por país

total_ciudades = cantidad_registros[cantidad_registros >= 30].index # Seleccionar países con al menos 30 registros
df_filtrados = df_validos[df_validos['country'].isin(total_ciudades)]

agrupar_datos = [group['daily_vaccinations'].values for name, group in df_filtrados.groupby('country')] # Agrupar los datos por país para ANOVA
resultado = f_oneway(*agrupar_datos) # Prueba ANOVA

# Mostrar resultados
print("Estadístico F:", resultado.statistic)
print("Valor p:", resultado.pvalue)

# Interpretación
if resultado.pvalue < 0.05: #Comparar con nivel de significancia
    print("Se rechaza la hipótesis nula: hay diferencias significativas entre países.")
else:
    print("No se rechaza la hipótesis nula: no hay diferencias significativas.")
```
