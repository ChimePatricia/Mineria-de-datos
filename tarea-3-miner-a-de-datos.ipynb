```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Cargar la base de datos
covid_file_path = '/kaggle/input/covid-world-vaccination-progress/country_vaccinations.csv'
df = pd.read_csv(covid_file_path)

# Convertir la columna de fecha a tipo datetime
df["date"] = pd.to_datetime(df["date"])
```

```python
#GRAFICO DE BARRAS
#Objetivo: Hacer el top de los 10 paises que han aplicado mas vacunas

# Agrupar por país y sumar vacunaciones totales
df_grouped = df.groupby("country")["total_vaccinations"].sum().sort_values(ascending=False).head(10)

# Gráfico de barras
plt.figure(figsize=(12, 6))
df_grouped.plot(kind="bar", color="skyblue")
plt.title("Top 10 Países con Más Vacunaciones Totales")
plt.ylabel("Total de Vacunaciones")
plt.xlabel("País")
plt.xticks(rotation=45)
plt.show()
```

```python
# DIAGRAMA PASTEL
# Objetivo: Mostrar la distribucion de vacunaciones por pais.

plt.figure(figsize=(8, 8))
df_grouped.plot(kind="pie", autopct="%1.1f%%", colormap="coolwarm")
plt.title("Distribución de Vacunaciones en el Top 10 Países")
plt.ylabel("")  # Ocultar el label de "y"
plt.show()
```
