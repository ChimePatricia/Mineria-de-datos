Objetivo:
Predecir si un país, en una fecha determinada, tuvo una vacunación diaria “alta” o “baja” comparado con el resto de los datos, usando información como:
Total de vacunas aplicadas
Personas con al menos una dosis
Personas totalmente vacunadas
